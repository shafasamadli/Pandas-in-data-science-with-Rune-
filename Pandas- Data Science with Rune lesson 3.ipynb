```python
# Hesablamalar apardiqdan sonra onu yeni sutuna yazib data sete elave etmek ucun
data['Data_new'] = data['Low'] - data['Close'] 
```

```python
data
```

```python
# select values, result will be boolean
data['Data_new'] > 1
```

```python
# xususile hansi deyerlerin 1-den kicik oldugunu gormek ucun 
data[data['Data_new'] < 1]
```

```python
# Categorize and group by 
# In this code part we just create new column and give boolean values to this column 
# Burada sadece dataya yeni bir column elave etdi ve data_new setrinde olan datalara sert verib onun esasinda yeni column(categorynin) icini doldurmus oldu
data['Category'] = data['Data_new'] > -0.5
data
```

```python
# Burada ise category columnunu mean deyerlerine gore qruplasdir dedim yeni datada false deyerler olanlarin(hansiki sert vermisdik ki data_new columnu -0.5den boyuk olsun) butun columnlar uzre ededi ortasi yeni mean deyerini hesablayib yazir, true deyerler ucun de eyni processi heyata kecirir
data.groupby('Category').mean()
```

```python
data['Category'].value_counts()
```

```python
(data['Data_new'] > -1).value_counts()
```
